#### Concatenate Numerical and Categorical dataframes into one dataframe called data. Split into X=features y=target.

```python
#concat cat with num 
categorical_df = categorical_df.reset_index(drop = True)
numerical_df = numerical_df.reset_index(drop = True)
data = pd.concat([categorical_df,numerical_df],axis = 1)
data
```

```python
#X, Y Split

X = data.drop(['total_claim_amount'], axis = 1)
y = data['total_claim_amount']
```

## 1. In this final lab, we will model our data. Import sklearn train_test_split and separate the data.

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, random_state=42)
```

```python
X_test_num = X_test.select_dtypes(include=np.number)
X_test_cat = X_test.select_dtypes(include=np.object)
```

## 3.Use X_train_num to fit scalers. Transform BOTH X_train_num and X_test_num.

```python
X_test_normalized = train_transformer.transform(X_test_num)
print(X_test_normalized.shape)
X_test_normalized
X_test_minmax = pd.DataFrame(X_test_normalized, columns=X_test_num.columns)
X_test_minmax
```

## 4. Encode the categorical variables X_train_cat and X_test_cat (See the hint below for encoding categorical data!!!)

```python
#split train in ohe/ordinal

X_train_cat_ohe = X_train_cat[['state', 'marital_status','policy', 'renew_offer_type', 'sales_channel','vehicle_class','response']]
X_train_cat_ordinal = X_train_cat[['employmentstatus','vehicle_size','education']]
```

```python
#split test in ohe/ordinal

X_test_cat_ohe = X_test_cat[['state', 'marital_status','policy', 'renew_offer_type', 'sales_channel','vehicle_class','response']]
X_test_cat_ordinal = X_test_cat[['employmentstatus','vehicle_size','education']]
```

```python
#encode train in ohe


encoder = OneHotEncoder(drop='first').fit(X_train_cat_ohe)
encoded_train = encoder.transform(X_train_cat_ohe).toarray()
cols = encoder.get_feature_names_out(input_features=X_train_cat_ohe.columns)
onehot_encoded_train = pd.DataFrame(encoded_train, columns=cols)
X_train_cat_encoded = onehot_encoded_train
```

```python
#encode test in ohe

encoded_test = encoder.transform(X_test_cat_ohe).toarray()
cols = encoder.get_feature_names_out(input_features=X_test_cat_ohe.columns)
onehot_encoded_test = pd.DataFrame(encoded_test, columns=cols)
X_test_cat_encoded = onehot_encoded_test
```

```python
#encode train in ordinal

X_train_cat_ordinal['employmentstatus'] = X_train_cat_ordinal['employmentstatus'].map({'Employed':1, 'Unemployed':0})
X_train_cat_ordinal['vehicle_size'] = X_train_cat_ordinal['vehicle_size'].map({'Small':0, 'Medsize':0.5, 'Large':1 })
X_train_cat_ordinal['education'] = X_train_cat_ordinal['education'].map({'High School or Below':0, 'College':0.25, 'Bachelor':0.5, 'master_&_dr':1})
```

```python
#encode test in ordinal

X_test_cat_ordinal['employmentstatus'] = X_test_cat_ordinal['employmentstatus'].map({'Employed':1, 'Unemployed':0})
X_test_cat_ordinal['vehicle_size'] = X_test_cat_ordinal['vehicle_size'].map({'Small':0, 'Medsize':0.5, 'Large':1 })
X_test_cat_ordinal['education'] = X_test_cat_ordinal['education'].map({'High School or Below':0, 'College':0.25, 'Bachelor':0.5, 'master_&_dr':1})
```

```python
#Concat Cat train ordinal+encoded

X_train_cat_ordinal = X_train_cat_ordinal.reset_index(drop = True)
X_train_cat_encoded = X_train_cat_encoded.reset_index(drop = True)


X_train_cat_transformed = pd.concat([X_train_cat_ordinal, X_train_cat_encoded ],axis=1)
X_train_cat_transformed
```

```python
#Concat Cat test ordinal+encoded

X_test_cat_ordinal = X_test_cat_ordinal.reset_index(drop = True)
X_test_cat_encoded = X_test_cat_encoded.reset_index(drop = True)

X_test_cat_transformed = pd.concat([X_test_cat_ordinal, X_test_cat_encoded ],axis=1)
X_test_cat_transformed
```

```python
#Concat Cat+num train

X_train_transform = pd.concat([X_train_cat_transformed, X_train_minmax], axis=1)
```

```python
#Concat Cat+num test

X_test_transform = pd.concat([X_test_cat_transformed, X_test_minmax], axis=1) 
```

##### Using metric = 'manhattan' and weigths = 'distance' improved the score.
##### Also, value of k changed from 8 to 11 (k=11) for best accuracy result.

```python
def predict(X_train_transform,y_train,model):
    if model=='linear':
        LR=LinearRegression().fit(X_train_transform,y_train)
        predictions = LR.predict(X_train_transform)
        print ('The r2 score is: ', r2_score(y_train,predictions))
    elif model=='kneighbors':
        scores = []
        k_val=[]
        for k in range(2,25):
            model = KNeighborsRegressor(n_neighbors=k,metric='manhattan',weights='distance')
            model.fit(X_train_transform, y_train)
            scores.append(model.score(X_test_transform, y_test))
            k_val.append(k)
        k_val=k_val[scores.index(max(scores))]
        print ('The optimum value of k is: ', k_val)
        print ('The accuracy score is: ', max(scores))  #just returning the best accuracy score 
    else:
        return 'Enter a valid model!'
```
